```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style='whitegrid')
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
print(iris.feature_names)
```

```python
print(iris.keys())
```

```python
df = pd.DataFrame(iris.data)
```

```python
df.shape
```

```python
type(iris.target)
```

```python
iris.target_names
```

```python
df['target'] = pd.Series(iris.target)
```

```python
df.shape
```

```python
df.columns = list(iris.feature_names)+list(['target'])
df.columns
```

```python
df.head()
```

```python
color_dict = {0:'red',1:'blue',2:'green'}
colors = []
for item in df.target.values:
    colors.append(color_dict[item])

fig, axes = plt.subplots(2, 2)
fig.set_size_inches([10,10])
axes[0,0].scatter(df['sepal length (cm)'],df['petal length (cm)'],c=colors)
axes[0,1].scatter(df['sepal length (cm)'],df['sepal width (cm)'],c=colors)
axes[1,0].scatter(df['petal length (cm)'],df['petal width (cm)'],c=colors)
axes[1,1].scatter(df['sepal width (cm)'],df['petal width (cm)'],c=colors)

plt.show()
```
